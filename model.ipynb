```python
import tensorflow as tf
import tflearn
from tflearn.layers.core import input_data, dropout, fully_connected
from tflearn.layers.conv import conv_2d, max_pool_2d
from tflearn.layers.estimator import regression
import numpy as np
import os
```

```python
IMG_SIZE = 64
MODEL_NAME = 'Model1_1'
```

```python
train_data = np.load('train.npy')
test_data = np.load('test.npy')
```

```python
train = train_data[:-500]
val = train_data[-500:]
```

```python
X = np.array([i[0] for i in train]).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
Y = np.array([i[1] for i in train])

val_x = np.array([i[0] for i in val]).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
val_y = np.array([i[1] for i in val])
```

```python
tf.reset_default_graph()
```

```python
#model
convnet = input_data(shape=[None, IMG_SIZE, IMG_SIZE, 1], name='input')
convnet = conv_2d(convnet, nb_filter=32, filter_size=5, activation='relu')
convnet = max_pool_2d(convnet, 5)
convnet = conv_2d(convnet, nb_filter=64, filter_size=5, activation='relu')
convnet = max_pool_2d(convnet, 5)
convnet = conv_2d(convnet, nb_filter=128, filter_size=5, activation='relu')
convnet = max_pool_2d(convnet, 5)
convnet = conv_2d(convnet, nb_filter=64, filter_size=5, activation='relu')
convnet = max_pool_2d(convnet, 5)
convnet = conv_2d(convnet, nb_filter=32, filter_size=5, activation='relu')
convnet = max_pool_2d(convnet, 5)

convnet = fully_connected(convnet, 1024, activation='relu')
convnet = dropout(convnet, 0.6)

convnet = fully_connected(convnet, 2, activation='softmax')
convnet = regression(convnet, name='targets')

model = tflearn.DNN(convnet, tensorboard_dir='log')
```

```python
model.fit({'input': X}, {'targets': Y}, n_epoch=10, validation_set=({'input': val_x}, {'targets': val_y}), 
    snapshot_step=500, show_metric=True, run_id=MODEL_NAME)
model.save(MODEL_NAME)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig=plt.figure(figsize=(10,10))

for num,data in enumerate(test_data[120:132]):
    # cat: [1,0]
    # dog: [0,1]
    
    img_num = data[1]
    img_data = data[0]
    
    y = fig.add_subplot(3,4,num+1)
    orig = img_data
    data = img_data.reshape(IMG_SIZE,IMG_SIZE,1)
    #model_out = model.predict([data])[0]
    model_out = model.predict([data])[0]
    
    if np.argmax(model_out) == 1: str_label='Dog'
    else: str_label='Cat'
        
    y.imshow(orig,cmap='gray')
    plt.title(str_label)
    y.axes.get_xaxis().set_visible(False)
    y.axes.get_yaxis().set_visible(False)
plt.show()
```
